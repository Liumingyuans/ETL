# http://feebee.com.tw  價格下載(By抓取的資料)

```python
#http://feebee.com.tw
import requests
import re
import time
from bs4 import BeautifulSoup as bs
for line in open('Lotions.txt'):
    a= line.split('|')[2]+line.split('|')[3]
    #a= line.split('|')[3]
    b = re.sub(r'[a-zA-Z]|-|．|(|)','',a) 
    print '商品名稱 '+b
    res=requests.get('http://feebee.com.tw/s/?q={}'.format(b))
    
    soup=bs(res.text)
    for i in range(1,6):
        try:
            print soup.select('.fb-u h2')[i].text,soup.select('.source span')[i].text,soup.select('.price em')[i].text
        except IndexError:
            print a+'--null--'
            pass
    print '---------------------------------------------------------------------------------------'
    #time.sleep(3)
```

# http://http://ezprice.com.tw/  價格下載(By抓取的資料)

```python
#http://http://ezprice.com.tw/
import requests
import re
import time
from bs4 import BeautifulSoup as bs
for line in open('price/ALL.csv'):
    f = open('price.csv', 'a+')
    a= line.split('|')[2].split(' ')[0]+' '+line.split('|')[3].split('(')[0]
    b= line.split('|')[0]
    if line.split('|')[1]=="[停產]":              #切割檔案第三項如果為[停產]則不抓資料
        print '[停產]商品名稱: '+a
        f.write(b+'|')
        f.write('[停產]商品名稱: '+a)
    else:
        f.write(b+'|')                            #ID(primary)
        print '商品名稱: '+a
        f.write('商品名稱: '+a+'|') 
        try:
            res=requests.get('http://ezprice.com.tw/s/{}/price/'.format(a))
        except :
            time.sleep(5)
            res=requests.get('http://ezprice.com.tw/s/{}/price/'.format(a))

        soup=bs(res.text)
        for i in range(0,3):
            try:
                #print soup.select('.srch_price')[i].text.strip(),soup.select('.srch_pf')[i].text.strip(),\
                #          soup.select('.srch_pdname a')[i].text.strip()
                f.write(soup.select('.srch_price')[i].text.strip().encode("utf-8")+'|')          #價錢寫入
                f.write(soup.select('.srch_pf')[i].text.strip().encode("utf-8")+'|')             #網站(ex:Yahoo超級商城)
                f.write(soup.select('.srch_pdname a')[i].text.strip().encode("utf-8")+'|')       #產品名稱
                f.write(soup.select('.srch_pdname a')[i]['href'].strip().encode("utf-8")+'|') 
            except IndexError:
                #print a+'--null--'
                f.write('$0|')
                f.write('NA|')
                f.write('NA|')
                f.write('NA|')
                pass
    f.write('\n')
    print '------------------------------------------------'
    time.sleep(1)
f.close()
```
